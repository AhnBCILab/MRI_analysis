```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
raw_data = pd.read_csv('output.csv')
raw_data.describe()
```

```python
hipo_left = np.array([raw_data[raw_data['Group'] == 'CN']['Left-Hippocampus'].mean(),
             raw_data[raw_data['Group'] == 'MCI']['Left-Hippocampus'].mean(),
             raw_data[raw_data['Group'] == 'AD']['Left-Hippocampus'].mean()])
hipo_right = np.array([raw_data[raw_data['Group'] == 'CN']['Right-Hippocampus'].mean(),
              raw_data[raw_data['Group'] == 'MCI']['Right-Hippocampus'].mean(),
              raw_data[raw_data['Group'] == 'AD']['Right-Hippocampus'].mean()])
hipo_err = np.array([(raw_data[raw_data['Group'] == 'CN']['Left-Hippocampus'] + raw_data[raw_data['Group'] == 'CN']['Right-Hippocampus']).std(),
            (raw_data[raw_data['Group'] == 'MCI']['Left-Hippocampus'] + raw_data[raw_data['Group'] == 'MCI']['Right-Hippocampus']).std(),
            (raw_data[raw_data['Group'] == 'AD']['Left-Hippocampus'] + raw_data[raw_data['Group'] == 'AD']['Right-Hippocampus']).std()])
```

```python
amyg_left = np.array([raw_data[raw_data['Group'] == 'CN']['Left-Amygdala'].mean(),
             raw_data[raw_data['Group'] == 'MCI']['Left-Amygdala'].mean(),
             raw_data[raw_data['Group'] == 'AD']['Left-Amygdala'].mean()])
amyg_right = np.array([raw_data[raw_data['Group'] == 'CN']['Right-Amygdala'].mean(),
              raw_data[raw_data['Group'] == 'MCI']['Right-Amygdala'].mean(),
              raw_data[raw_data['Group'] == 'AD']['Right-Amygdala'].mean()])
amyg_err = np.array([(raw_data[raw_data['Group'] == 'CN']['Left-Amygdala'] + raw_data[raw_data['Group'] == 'CN']['Right-Amygdala']).std(),
            (raw_data[raw_data['Group'] == 'MCI']['Left-Amygdala'] + raw_data[raw_data['Group'] == 'MCI']['Right-Amygdala']).std(),
            (raw_data[raw_data['Group'] == 'AD']['Left-Amygdala'] + raw_data[raw_data['Group'] == 'AD']['Right-Amygdala']).std()])
```

```python
latr_left = np.array([raw_data[raw_data['Group'] == 'CN']['Left-Inf-Lat-Vent'].mean(),
             raw_data[raw_data['Group'] == 'MCI']['Left-Inf-Lat-Vent'].mean(),
             raw_data[raw_data['Group'] == 'AD']['Left-Inf-Lat-Vent'].mean()])
latr_right = np.array([raw_data[raw_data['Group'] == 'CN']['Right-Inf-Lat-Vent'].mean(),
              raw_data[raw_data['Group'] == 'MCI']['Right-Inf-Lat-Vent'].mean(),
              raw_data[raw_data['Group'] == 'AD']['Right-Inf-Lat-Vent'].mean()])
latr_err = np.array([(raw_data[raw_data['Group'] == 'CN']['Left-Inf-Lat-Vent'] + raw_data[raw_data['Group'] == 'CN']['Right-Inf-Lat-Vent']).std(),
            (raw_data[raw_data['Group'] == 'MCI']['Left-Inf-Lat-Vent'] + raw_data[raw_data['Group'] == 'MCI']['Right-Inf-Lat-Vent']).std(),
            (raw_data[raw_data['Group'] == 'AD']['Left-Inf-Lat-Vent'] + raw_data[raw_data['Group'] == 'AD']['Right-Inf-Lat-Vent']).std()])
```

```python
index = [1, 1.6, 2.2]

fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(12,5))

axes[0].bar(index, hipo_left, color='r', alpha = .5, width=.3)
axes[0].bar(index, hipo_right, color='y', alpha = .5, bottom=hipo_left, width=.3)
_, c, _ = axes[0].errorbar(index, hipo_left + hipo_right, yerr=hipo_err, capsize=0, lolims=True, ls='None', c='k')
c[0].set_marker('_')
c[0].set_markersize(10)

axes[0].set_xlabel('Group')
axes[0].set_ylabel(r'Volume ($mm^3$)')
axes[0].set_title('Hippocampus')
plt.sca(axes[0])
plt.xticks(index, ['NC', 'MCI', 'AD'])
plt.legend(['Left', 'Right'], loc='lower right')

data = [hipo_left, hipo_right, hipo_err]

x1, x2 = 1, 1.6
y, h = max((data[0][0] + data[1][0] + data[2][0]).max(), (data[0][1] + data[1][1] + data[2][1]).max()) + 300, 100
plt.plot([x1, x1, x2, x2], [y, y+h, y+h, y], lw=1.5, color='k')
plt.text((x1 + x2) * .5, y+h, "*", ha='center', va='bottom', color='k')

x1, x2 = 1.6, 2.2
y, h = max((data[0][2] + data[1][2] + data[2][2]).max(), (data[0][1] + data[1][1] + data[2][1]).max()) + 300, 100
plt.plot([x1, x1, x2, x2], [y, y+h, y+h, y], lw=1.5, color='k')
plt.text((x1 + x2) * .5, y+h, "*", ha='center', va='bottom', color='k')

x1, x2 = 1, 2.2
y, h = max((data[0][2] + data[1][2] + data[2][2]).max(), (data[0][0] + data[1][0] + data[2][0]).max()) + 800, 100
plt.plot([x1, x1, x2, x2], [y, y+h, y+h, y], lw=1.5, color='k')
plt.text((x1 + x2) * .5, y+h, "*", ha='center', va='bottom', color='k')

axes[1].bar(index, amyg_left, color='r', alpha = .5, width=.3)
axes[1].bar(index, amyg_right, color='y', alpha = .5, bottom=amyg_left, width=.3)
_, c, _ = axes[1].errorbar(index, amyg_left + amyg_right, yerr=amyg_err, capsize=0, lolims=True, ls='None', c='k')
c[0].set_marker('_')
c[0].set_markersize(10)

axes[1].set_xlabel('Group')
#axes[1].set_ylabel('Volume (mm^3)')
axes[1].set_title('Amygdala')
plt.sca(axes[1])
plt.xticks(index, ['NC', 'MCI', 'AD'])
plt.legend(['Left', 'Right'], loc='lower right')

data = [amyg_left, amyg_right, amyg_err]

x1, x2 = 1, 1.6
y, h = max((data[0][0] + data[1][0] + data[2][0]).max(), (data[0][1] + data[1][1] + data[2][1]).max()) + 200, 50
plt.plot([x1, x1, x2, x2], [y, y+h, y+h, y], lw=1.5, color='k')
plt.text((x1 + x2) * .5, y+h, "*", ha='center', va='bottom', color='k')

x1, x2 = 1.6, 2.2
y, h = max((data[0][2] + data[1][2] + data[2][2]).max(), (data[0][1] + data[1][1] + data[2][1]).max()) + 200, 50
plt.plot([x1, x1, x2, x2], [y, y+h, y+h, y], lw=1.5, color='k')
plt.text((x1 + x2) * .5, y+h, "*", ha='center', va='bottom', color='k')

x1, x2 = 1, 2.2
y, h = max((data[0][2] + data[1][2] + data[2][2]).max(), (data[0][0] + data[1][0] + data[2][0]).max()) + 400, 50
plt.plot([x1, x1, x2, x2], [y, y+h, y+h, y], lw=1.5, color='k')
plt.text((x1 + x2) * .5, y+h, "*", ha='center', va='bottom', color='k')

axes[2].bar(index, latr_left, color='r', alpha = .5, width=.3)
axes[2].bar(index, latr_right, color='y', alpha = .5, bottom=latr_left, width=.3)
_, c, _ = axes[2].errorbar(index, latr_left + latr_right, yerr=latr_err, capsize=0, lolims=True, ls='None', c='k')
c[0].set_marker('_')
c[0].set_markersize(10)

axes[2].set_xlabel('Group')
#axes[2].set_ylabel('Volume (mm^3)')
axes[2].set_title('Inferior Lateral Ventricle')
plt.sca(axes[2])
plt.xticks(index, ['NC', 'MCI', 'AD'])
plt.legend(['Left', 'Right'], loc='lower right')

data = [latr_left, latr_right, latr_err]

x1, x2 = 1, 1.6
y, h = max((data[0][0] + data[1][0] + data[2][0]).max(), (data[0][1] + data[1][1] + data[2][1]).max()) + 1000, 100
plt.plot([x1, x1, x2, x2], [y, y+h, y+h, y], lw=1.5, color='k')
plt.text((x1 + x2) * .5, y+h, "*", ha='center', va='bottom', color='k')

x1, x2 = 1.6, 2.2
y, h = max((data[0][2] + data[1][2] + data[2][2]).max(), (data[0][1] + data[1][1] + data[2][1]).max()) + 300, 100
plt.plot([x1, x1, x2, x2], [y, y+h, y+h, y], lw=1.5, color='k')
plt.text((x1 + x2) * .5, y+h, "*", ha='center', va='bottom', color='k')

x1, x2 = 1, 2.2
y, h = max((data[0][2] + data[1][2] + data[2][2]).max(), (data[0][0] + data[1][0] + data[2][0]).max()) + 700, 100
plt.plot([x1, x1, x2, x2], [y, y+h, y+h, y], lw=1.5, color='k')
plt.text((x1 + x2) * .5, y+h, "*", ha='center', va='bottom', color='k')

fig.tight_layout()
fig.savefig('result2.png', dpi=300)
plt.show()
```
