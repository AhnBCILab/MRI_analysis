```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
raw_data = pd.read_csv('ADNI_dataset.csv')
raw_data = raw_data[raw_data["CDR"] != "None"]
for column in raw_data.columns[32:]:
    raw_data = raw_data[raw_data[column] != "None"]
raw_data.info()
```

```python
def normalize(data):
    from sklearn.preprocessing import StandardScaler
    std_normalizer = StandardScaler()
    std_normalizer.fit(data)
    norm_data = std_normalizer.transform(data)
    return norm_data
```

```python
roi_data = raw_data.loc[:, 'BrainSeg':]

CDR_data = pd.DataFrame(map(float, raw_data["CDR"]))
CDR_data.columns = ["CDR"]
roi_data = pd.concat([CDR_data, roi_data], axis=1)
```

```python
NC = roi_data[roi_data["CDR"] == 0.0].dropna()
MCI = roi_data[roi_data["CDR"] == 0.5].dropna()
AD = roi_data[roi_data["CDR"] >= 1.0].dropna()

NC_norm = pd.DataFrame(normalize(NC), columns=roi_data.columns, index=(NC.index.values))
MCI_norm = pd.DataFrame(normalize(MCI), columns=roi_data.columns, index=(MCI.index.values))
AD_norm = pd.DataFrame(normalize(AD), columns=roi_data.columns, index=(AD.index.values))
```

```python
from scipy.stats import f_oneway
from scipy.stats import ttest_ind

fig = plt.figure(figsize=(30, 200))
ax = {}
p_all = {}
i = 0
for col in roi_data.columns:
    if col == "CDR":
        continue
    f, p = f_oneway(NC_norm[col], MCI_norm[col], AD_norm[col])
    tp_NM = ttest_ind(NC_norm[col], MCI_norm[col])
    tp_MA = ttest_ind(MCI_norm[col], AD_norm[col])
    tp_NA = ttest_ind(NC_norm[col], AD_norm[col])
    p_all[i] = p
    
    ax[i] = fig.add_subplot(23, 3, i + 1)
    if p < 0.001:
        ax[i].set_title(col + " ***", fontsize=20)
    elif p < 0.01:
        ax[i].set_title(col + " **", fontsize=20)
    elif p < 0.05:
        ax[i].set_title(col + " *", fontsize=20)
    else:
        ax[i].set_title(col, fontsize=20)
    ax[i].boxplot([NC_norm[col], MCI_norm[col], AD_norm[col]])
    ax[i].set_ylim(-4, 5)
    ax[i].set_xticklabels(["NC", "MCI", "AD"])
    
    ax[i].text(0.5, -4.7, "p-value = " + ("%.2f" % p))
    if tp_NM.pvalue < 0.05:
        ax[i].plot([1, 2], [3.8, 3.8])
        if tp_NM.pvalue < 0.001:
            ax[i].text(1.5, 4, "***")
        elif tp_NM.pvalue < 0.01:
            ax[i].text(1.5, 4, "**")
        elif tp_NM.pvalue < 0.05:
            ax[i].text(1.5, 4, "*")
    if tp_MA.pvalue < 0.05:
        ax[i].plot([2, 3], [4.2, 4.2])
        if tp_NM.pvalue < 0.001:
            ax[i].text(2.5, 4.4, "***")
        elif tp_NM.pvalue < 0.01:
            ax[i].text(2.5, 4.4, "**")
        elif tp_NM.pvalue < 0.05:
            ax[i].text(2.5, 4.4, "*")
    if tp_NA.pvalue < 0.05:
        ax[i].plot([1, 3], [4.7, 4.7])
        if tp_NM.pvalue < 0.001:
            ax[i].text(3.5, 4.9, "***")
        elif tp_NM.pvalue < 0.01:
            ax[i].text(3.5, 4.9, "**")
        elif tp_NM.pvalue < 0.05:
            ax[i].text(3.5, 4.9, "*")
    
    i += 1
fig.show()
plt.savefig('p-value result dataset within normalize.png')
```

```python
for index, p in enumerate(p_all.values()):
    if p <= 0.05:
        print(roi_data.columns[index + 1])
```

```python
def make_lhipo_df(data, cdr):
    con_data = pd.DataFrame(data['Left-Hippocampus'] / data['BrainSeg'], 
                      columns=["Hippocampus Rate"])
    con_data.reset_index(drop=True, inplace=True)
    return pd.concat([pd.DataFrame(list(np.full(data.shape[0], cdr)), columns=["CDR"]), con_data], axis=1)

def make_rhipo_df(data, cdr):
    con_data = pd.DataFrame(data['Right-Hippocampus'] / data['BrainSeg'], 
                      columns=["Hippocampus Rate"])
    con_data.reset_index(drop=True, inplace=True)
    return pd.concat([pd.DataFrame(list(np.full(data.shape[0], cdr)), columns=["CDR"]), con_data], axis=1).dropna()

NC_l_hipo_rate = make_lhipo_df(NC, 0)
MCI_l_hipo_rate = make_lhipo_df(MCI, 0.5)
AD_l_hipo_rate = make_lhipo_df(AD, 1)
L_hipo_rate = pd.concat([NC_l_hipo_rate, MCI_l_hipo_rate, AD_l_hipo_rate])
L_hipo_rate_list = [NC_l_hipo_rate, MCI_l_hipo_rate, AD_l_hipo_rate]

NC_r_hipo_rate = make_rhipo_df(NC, 0)
MCI_r_hipo_rate = make_rhipo_df(MCI, 0.5)
AD_r_hipo_rate = make_rhipo_df(AD, 1)
R_hipo_rate = pd.concat([NC_r_hipo_rate, MCI_r_hipo_rate, AD_r_hipo_rate])
R_hipo_rate_list = [NC_r_hipo_rate, MCI_r_hipo_rate, AD_r_hipo_rate]

NC_hipo_rate = NC_l_hipo_rate + NC_r_hipo_rate
MCI_hipo_rate = MCI_l_hipo_rate + MCI_r_hipo_rate
AD_hipo_rate = AD_l_hipo_rate + AD_r_hipo_rate
Hipo_rate = pd.concat([NC_hipo_rate, MCI_hipo_rate, AD_hipo_rate])
Hipo_rate_list = [NC_hipo_rate, MCI_hipo_rate, AD_hipo_rate]
```

```python
from scipy.stats import f_oneway
from scipy.stats import ttest_ind

fig = plt.figure(figsize=(20, 8))
ax = {}
p_all = {}
i = 0
names = ["Left Hippocampus Rate", "Right Hippocampus Rate", "Whole Hippocampus Rate"]

# Normal Cognitive Left Hippocampus Rate
for (NC_h, MCI_h, AD_h) in [L_hipo_rate_list, R_hipo_rate_right, Hipo_rate_list]:
    NC_h = NC_h.iloc[:, 1]
    MCI_h = MCI_h.iloc[:, 1]
    AD_h = AD_h.iloc[:, 1]
    
    f, p = f_oneway(NC_h, MCI_h, AD_h)
    tp_NM = ttest_ind(NC_h, MCI_h, equal_var=False)
    tp_MA = ttest_ind(MCI_h, AD_h, equal_var=False)
    tp_NA = ttest_ind(NC_h, AD_h, equal_var=False)
    p_all[i] = p
    
    ax[i] = fig.add_subplot(1, 3, i + 1)
    ax[i].set_title(names[i], fontsize=20)
    ax[i].boxplot([NC_h, MCI_h, AD_h])
    # ax[i].set_ylim(-4, 5)
    ax[i].set_xticklabels(["NC", "MCI", "AD"])
    
    y_bottom, y_top = ax[i].get_ylim()
    ax[i].text(0.5, y_bottom, "p-value = " + ("%.3f" % p))
    print(tp_NM.pvalue, tp_MA.pvalue, tp_NA.pvalue)
    if tp_NM.pvalue < 0.05:
        ax[i].plot([1, 2], [y_top, y_top])
        if tp_NM.pvalue < 0.001:
            ax[i].text(1.5, y_top, "***")
        elif tp_MA.pvalue < 0.01:
            ax[i].text(1.5, y_top, "**")
        elif tp_NA.pvalue < 0.05:
            ax[i].text(1.5, y_top, "*")
    if tp_MA.pvalue < 0.05:
        ax[i].plot([2, 3], [y_top, y_top])
        if tp_NM.pvalue < 0.001:
            ax[i].text(2.5, y_top, "***")
        elif tp_MA.pvalue < 0.01:
            ax[i].text(2.5, y_top, "**")
        elif tp_NA.pvalue < 0.05:
            ax[i].text(2.5, y_top, "*")
    if tp_NA.pvalue < 0.05:
        ax[i].plot([1, 3], [y_top, y_top])
        if tp_NM.pvalue < 0.001:
            ax[i].text(2, y_top, "***")
        elif tp_MA.pvalue < 0.01:
            ax[i].text(2, y_top, "**")
        elif tp_NA.pvalue < 0.05:
            ax[i].text(2, y_top, "*")
    
    i += 1
fig.show()
plt.savefig('Hippocampus rate p-value without normalize.png')
```
