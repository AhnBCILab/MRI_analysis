```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
raw_data = pd.read_csv('ADNI_dataset.csv')
raw_data = raw_data[raw_data["CDR"] != "None"]
for column in raw_data.columns[32:]:
    raw_data = raw_data[raw_data[column] != "None"]
raw_data.info()
```

```python
roi_data = raw_data.loc[:, 'BrainSeg':]

from sklearn.preprocessing import StandardScaler
std_normalizer = StandardScaler()
std_normalizer.fit(roi_data)
norm_data = std_normalizer.transform(roi_data)
norm_data = pd.DataFrame(norm_data, columns=roi_data.columns, index=(roi_data.index.values))

CDR_data = pd.DataFrame(map(float, raw_data["CDR"]))
CDR_data.columns = ["CDR"]
norm_data = pd.concat([CDR_data, norm_data], axis=1)
norm_data.describe()
```

```python
NC = norm_data[norm_data["CDR"] == 0.0]
MCI = norm_data[norm_data["CDR"] == 0.5]
AD = norm_data[norm_data["CDR"] >= 1.0]

NC = NC.dropna()
MCI = MCI.dropna()
AD = AD.dropna()
```

```python
NC.describe()
```

```python
MCI.describe()
```

```python
AD.describe()
```

```python
from scipy.stats import f_oneway
from scipy.stats import ttest_ind

fig = plt.figure(figsize=(30, 200))
ax = {}
p_all = {}
i = 0
for col in norm_data.columns:
    if col == "CDR":
        continue
    f, p = f_oneway(NC[col], MCI[col], AD[col])
    tp_NM = ttest_ind(NC[col], MCI[col], equal_var=False)
    tp_MA = ttest_ind(MCI[col], AD[col], equal_var=False)
    tp_NA = ttest_ind(NC[col], AD[col], equal_var=False)
    p_all[i] = p
    
    ax[i] = fig.add_subplot(23, 3, i + 1)
    ax[i].set_title(col, fontsize=20)
    ax[i].boxplot([NC[col], MCI[col], AD[col]])
    ax[i].set_ylim(-4, 5)
    ax[i].set_xticklabels(["NC", "MCI", "AD"])
    
    ax[i].text(0.5, -4.7, "p-value = " + ("%.2f" % p))
    if tp_NM.pvalue < 0.05:
        ax[i].plot([1, 2], [4.2, 4.2])
        print("NM / %s" % col)
    if tp_MA.pvalue < 0.05:
        ax[i].plot([2, 3], [4.5, 4.5])
        print("MA / %s" % col)
    if tp_NA.pvalue < 0.05:
        ax[i].plot([1, 3], [4.8, 4.8])
        print("NA / %s" % col)
    
    i += 1
fig.show()
plt.savefig('p-value result dataset.png')
```

```python
for index, p in enumerate(p_all.values()):
    if p <= 0.05:
        print(norm_data.columns[index + 1])
```
